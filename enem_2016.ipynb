#### Agora que nao temos nenhum dado nulo nas colunas de nota podemos encontrar os 20 melhores colocados
##### Para isso vamos criar uma nova coluna chamada 'MEDIA_P' que fara a media ponderada de cada candidato


```python
zeros = np.linspace(0,0,9860)
df_presentes['MEDIA_P'] = zeros
df_presentes.head()
```

```python
#calculando os pesos e gerando a media
df_presentes['MEDIA_P'] = (df_presentes['NU_NOTA_CN'].apply(lambda x: x*2) + df_presentes['NU_NOTA_MT'].apply(lambda x: x*3) +
    df_presentes['NU_NOTA_CH'].apply(lambda x: x*1) + df_presentes['NU_NOTA_LC'].apply(lambda x: x*1.5) + df_presentes['NU_NOTA_REDACAO'].apply(lambda x: x*3))//10.5
    
```

```python
#ordenar pela MEDIA_P
df_presentes = df_presentes.sort_values(by=['MEDIA_P'],ascending=False) 
```

## Agora todas as colunas possuem valores numericos, podemos partir para nosso modelo de regressão.
### Vamos considerar as colunas com maior correlação com 'NU_NOTA_MT'

```python
df_train = df.iloc[:, 43:55].drop(['TP_LINGUA', 'TP_STATUS_REDACAO'], axis=1)
df_train.head()
```

```python
x_train = df_train.iloc[:, :].drop('NU_NOTA_MT', axis=1)
y_train = df_train['NU_NOTA_MT']
```

```python
x_train.shape, y_train.shape
```

```python
# dados de teste
df_test = pd.read_csv('test.csv')

#selecionando as colunas uteis
cols = x_train.columns
df_test = df_test[cols]
df_test
```

```python
x_test = df_test.fillna(0)
x_test
```

# Modelo de Regressão

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)
```

```python
y_predict = lr.predict(x_test)
y_predict
```

### Usando df_train com split do scikitlearn

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.33, random_state=42)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

### LinearRegression

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)
```

```python
prediction = lr.predict(x_test)
prediction
```

```python
# MSE
from sklearn.metrics import mean_squared_error


mean_squared_error(y_test, prediction)
```

### GradientBoostRegressor

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gb = GradientBoostingRegressor()
gb.fit(x_train,y_train)
gb.score(x_train,y_train)
```

```python
prediction = gb.predict(x_test)
prediction
```

```python
mean_squared_error(y_test, prediction)
```

## BayesianRidge

```python
from sklearn.linear_model import BayesianRidge

br = BayesianRidge()
br = br.fit(x_train, y_train)

br.score(x_train, y_train)
```

```python
prediction = br.predict(x_test)
prediction
```

```python
print("Score BayesianRidge:", br.score(x_test, y_test))  #Return the coefficient of determination R^2 of the prediction.
```
